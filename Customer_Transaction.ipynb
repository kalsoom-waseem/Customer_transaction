**Adding a new column named Quantity and Making Transaction Product**

```python
df['quantity']=1
transaction_product=df.groupby(['basket_id','commodity'])['quantity'].sum().unstack().reset_index().fillna(0).set_index('basket_id')
transaction_product.reset_index(inplace=True)
display(transaction_product)
```

**Transform the data into 1 and 0 s**

```python
def encode(x):
  if x>=1:
    return 1
  else:
    return 0
transaction_product=transaction_product.applymap(encode)

print(transaction_product)
```

**Transform a new datset customer_product_freq**

```python
df['quantity']=1
customer_product_freq=df.groupby(['customer_id','commodity'])['quantity'].sum().unstack().reset_index().fillna(0).set_index('customer_id')
customer_product_freq.reset_index(inplace=True)
display(customer_product_freq)
```

****

**KNN Neighbors**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
n_df=pd.DataFrame(customer_product_freq)
y=n_df.pop('customer_id')
x=n_df

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
knn=KNeighborsClassifier(n_neighbors=5)

knn.fit(x,y)
```

```python
print(knn.score(x_test,y_test))
```

**Top 5 similar customers for each customer**

```python
index=knn.kneighbors(X=x_train, n_neighbors=5, return_distance=False)
ind=0
for i in range(len(x_train)):
  print('Customer: ',y[ind])
  for j in range(5):
    print(customer_product_freq.loc[index[i][j],'customer_id'])

  ind+=1
```

**Import Apriori**

```python
import mlxtend
from mlxtend import frequent_patterns
from mlxtend.frequent_patterns import apriori
```

```python
#transaction_product_copy=pd.DataFrame(transaction_product)
#transaction_product_copy.
transaction_product.set_index('basket_id',inplace=True)
```

**The itemsets(basket) having length more than 1 and minimum support of 5%**

```python
frequent_itemsets_plus=apriori(transaction_product, min_support= 0.05,use_colnames=True).sort_values('support',ascending=False).reset_index(drop=True)
frequent_itemsets_plus['length']= frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))
frequent_itemsets_plus
frequent_itemsets_plus[(frequent_itemsets_plus['length']>1)& frequent_itemsets_plus['support']>=0.05]
```

```python
frequent_itemsets_plus[frequent_itemsets_plus['support']>=0.02]
```

```python
from mlxtend.frequent_patterns import association_rules
```

**The association rules with minimum support of 2% and having lift more than 1.**

```python
association_rules(frequent_itemsets_plus,metric='lift',min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```
